```python
#extracting function
def extract_columns_from_dataframes(dataframes, columns):
    """
    Extract specified columns from a list of DataFrames.

    Parameters:
    - dataframes: List of DataFrames to extract columns from.
    - columns: List of column names to extract.

    Returns:
    - A new DataFrame containing the extracted columns from all DataFrames.
    """
    extracted_dataframes = []  # List to hold the extracted DataFrames

    for df in dataframes:
        # Extract the specified columns, ensuring they exist in the DataFrame
        extracted_df = df[columns].copy() if all(col in df.columns for col in columns) else pd.DataFrame()

        # Append the extracted DataFrame to the list
        extracted_dataframes.append(extracted_df)

    # Concatenate all extracted DataFrames into one
    combined_extracted_df = pd.concat(extracted_dataframes, ignore_index=True)

    return combined_extracted_df



```

```python
# Function to classify children based on age group and hours worked

def classify_on_track(row):
    age = row['Age of child']
    hours = row['Hours worked in past week']

    # Check age conditions and classify based on hours
    if 5 <= age <= 11:
        return 1 if hours >= 1 else 0
    elif 12 <= age <= 14:
        return 1 if hours >= 14 else 0
    elif 15 <= age <= 17:
        return 1 if hours >= 43 else 0
    else:
        return np.nan  # Return NaN for ages outside the specified range
```

```python
# Function to classify children based on age group and hours worked to identify child labour

def classify_on_track(row):
    age = row['Age of child']
    hours = row['Number of hours engaged in past week']

    # Classify child labour based on age-specific number of hours
    if 5 <= age <= 14:
        return 1 if hours >= 21 else 0  # 21 hours or more indicates child labour
    elif 15 <= age <= 17:
        return 1  # No limit on hours, any work indicates child labour
    else:
        return np.nan  # Return NaN for ages outside the specified range
```

```python
def classify_on_track(row):
    age = row['Age of child']
    water_hours = row.get('Hours spent fetching water in past week', 0)
    firewood_hours = row.get('Hours spent collecting firewood in past week', 0)
    other_hours = row.get('Number of hours engaged in past week', 0)

    # Aggregate total hours spent on household chores
    total_hours = water_hours + firewood_hours + other_hours

    # Classify child labour based on age and total hours
    if 5 <= age <= 14:
        return 1 if total_hours >= 21 else 0  # 21 hours or more indicates child labour
    elif 15 <= age <= 17:
        return 1  # No limit on hours, any engagement indicates child labour
    else:
        return np.nan  # Return NaN for ages outside the specified range
```

```python
# Function to export dataframe to Stata with value labels
def export_to_stata(df, file_name, value_labels):
    # Apply value labels to categorical columns
    for col, labels in value_labels.items():
        df[col] = df[col].cat.rename_categories(labels)

    # Export to Stata file
    df.to_stata(file_name, write_index=False)
```

```python
value_labels = {'Child_Labour_EA': {0: 'No Child Labour', 1: 'Child Labour'},
                'Child_Labour_HC': {0: 'No Child Labour', 1: 'Child Labour'}}

export_to_stata(df, 'output.dta', value_labels)
```
